## Finding and Droping the outliers

```python
num = df.select_dtypes(include='number')
def outliers(col):
    q1,q3 = np.percentile(df[col],[25,75])
    IQR = q3-q1
    lb = q1-(1.5*IQR)
    ub = q3+(1.5*IQR)
    outlier = df[(df[col]<lb)|(df[col]>ub)]
    return len(outlier)
for i in num:
    print(i,outliers(i)) 
```

### Segregating input and output values

```python
X = df[['input']]
y = df['output']
```

### Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

### Standardizing the input values in train and test

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_transform = scaler.fit_transform(X_train)
```

```python
X_test_transform = scaler.transform(X_test)
```

## Building the model 

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(X_train_transform,y_train)
```

### Predicting the values

```python
y_pred = le.predict(X_test_transform)
```

### Evaluating the model

```python
from sklearn import metrics

metrics.mean_absolute_error(y_test,y_pred)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
de = DecisionTreeRegressor()
de.fit(X_train_transform,y_train)
y_pred_de = de.predict(X_test_transform)
metrics.mean_absolute_error(y_test,y_pred_de)
```

### KNN regressor

```python
from sklearn.neighbors import KNeighborsRegressor
kn = KNeighborsRegressor()
kn.fit(X_train_transform,y_train)
y_pred_kn = kn.predict(X_test_transform)
metrics.mean_absolute_error(y_test,y_pred_kn)
```

### Support Vector Regressor

```python
from sklearn.svm import SVR
sv = SVR()
sv.fit(X_train_transform,y_train)
y_pred_sv = sv.predict(X_test_transform)
metrics.mean_absolute_error(y_test,y_pred_sv)
```

### Adaboost regressor

```python
from sklearn.ensemble import AdaBoostRegressor
ad = AdaBoostRegressor()
ad.fit(X_train_transform,y_train)
y_pred_ad = ad.predict(X_test_transform)
metrics.mean_absolute_error(y_test,y_pred_ad)
```

### Random Forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train_transform,y_train)
y_pred_rf = rf.predict(X_test_transform)
metrics.mean_absolute_error(y_test,y_pred_rf)
```

### Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(X_train_transform,y_train)
y_pred_gb = gb.predict(X_test_transform)
metrics.mean_absolute_error(y_test,y_pred_gb)
```
